```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon
```

```python
## uniform distribution
## Distribution is constant between loc and loc + scale.

num_trials = 10000
start = 0
width = 10

trials_uniform = uniform.rvs(size = num_trials, loc = start, scale = width)

trials_uniform
```

```python
np.count_nonzero(trials_uniform)
```

```python
trials_uniform.mean()
```

```python
##Histogram of trials

unif_n, unif_bins, unif_patches = plt.hist(trials_uniform, bins = 10)
plt.show()
```

```python
plt.scatter(x = range(num_trials), y = trials_uniform, s = 1)
plt.show()
```

```python
unif_n
```

```python
np.cumsum(unif_n)
```

```python
unif_bins
```

```python
plt.plot(unif_bins[1:], np.cumsum(unif_n)/num_trials)
plt.title('Uniform Cumulative Distribution Function')
plt.show()
```

```python
## Normal distribution
## arguments in rvs function:  loc = mean and scale = std dev

num_trials = 10000
mean = 50
sd = 10

trials_normal = norm.rvs(size = num_trials, loc = mean, scale = sd)

print('mean:', trials_normal.mean(), 'sd:', trials_normal.std())
```

```python
##Histogram of trials

norm_n, norm_bins, norm_patches = plt.hist(trials_normal, bins = 50)
plt.show()
```

```python
plt.scatter(x = range(num_trials), y = trials_normal, s = 1)
plt.show()
```

```python
plt.plot(norm_bins[1:], np.cumsum(norm_n)/num_trials)
plt.title('Normal Cumulative Distribution Function')
plt.show()
```

```python
## Expontential Distribution
## arguments in rvs function:  loc shifts distribution, scale = 1/lambda (also = mean)

## We will come back to exponential distribution in queuing models.
##  There, lambda is the rate of occurrences per unit of time or, alternatively worded
##          the expected time in between customers (say, 1 customer every half hour).
## 1/lambda is therefore the expected number of occurrences in the time interval (2 customers expected in one hour)

num_trials = 10000
lambda_parm = .5
shift = 0

trials_expon = expon.rvs(size = num_trials, loc = shift, scale = 1/lambda_parm)

print('mean:', trials_expon.mean())
```

```python
##Histogram of trials

expon_n, expon_bins, expon_patches = plt.hist(trials_expon, bins = 50)
plt.show()
```

```python
plt.scatter(x = range(num_trials), y = trials_expon, s = 1)
plt.show()
```

```python
plt.plot(expon_bins[1:], np.cumsum(expon_n)/num_trials)
plt.title('Exponential Cumulative Distribution Function')
plt.show()
```
